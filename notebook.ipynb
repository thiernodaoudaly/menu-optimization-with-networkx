<h2 style="text-align:center; color:blue">I. Description du problème</h2>
Dans notre projet, nous nous concentrons particulièremment sur la gestion des régimes alimentaires dans les environnements hôteliers et autres lieux hautement fréquentés. Notre objectif est de proposer une solution efficace et inclusive pour répondre aux besoins alimentaires variés des clients, en minimisant le nombre de plats proposés tout en assurant une diversité suffisante pour satisfaire ces exigences. 

Pour ce faire, nous prenons l'exemple de huit régimes alimentaires différents, chacun présentant des incompatibilités spécifiques :

- **le régime "Sans gluten" exclut les produits à base de blé, d'orge ou de seigle.**
- **le régime "Végétalien" exclut les produits d'origine animale tels que la viande, les œufs et le miel.**
- **le régime "Sans lactose" exclut les produits laitiers ou dérivés contenant du lactose.**
- **le régime "Sans fruits à coque" exclut les amandes, noisettes, noix, etc.**
- **le régime "Sans produits laitiers" exclut le lait, le yaourt, le fromage, etc.**
- **le régime "Sans œufs" exclut les œufs et les produits en contenant.**
- **le régime "Sans poisson" exclut le poisson et les produits à base de poisson.**
- **le régime "Sans viande rouge" exclut le bœuf, le porc, l'agneau, etc.**

Ces incompatibilités représentent les restrictions que chaque régime alimentaire implique, et elles doivent être prises en compte lors de la planification des repas ou des menus pour répondre aux besoins de chaque personne.

<h2 style="text-align:center; color:red">II. Technologies utilisées</h2>

Pour aborder le projet de gestion des régimes alimentaires dans les environnements hôteliers et autres lieux fréquentés, nous avons utilisé la bibliothèque ***NetworkX*** de Python.

***NetworkX*** est une bibliothèque Python incontournable pour la création, la manipulation et l'analyse de structures de graphes. Elle met à disposition une multitude d'outils pour la modélisation, la visualisation, la simulation et l'analyse des graphes. En particulier, ***NetworkX*** se distingue par son efficacité dans la résolution de problèmes d'optimisation combinatoire sur des graphes, notamment la coloration des graphes.

***La coloration des graphes*** est un problème classique en théorie des graphes et en recherche opérationnelle. Elle consiste à attribuer une couleur à chaque nœud d'un graphe de telle manière que deux nœuds adjacents aient des couleurs différentes. L'objectif principal est de minimiser le nombre total de couleurs utilisées pour colorier l'ensemble du graphe. Dans notre contexte, nous adaptons ce concept pour résoudre notre problème de gestion des régimes alimentaires. Plus précisément, nous utilisons la coloration des graphes pour déterminer le nombre minimum de plats nécessaires pour satisfaire tous les régimes alimentaires tout en évitant les incompatibilités entre eux.

En utilisant ***NetworkX***, nous pouvons créer un graphe non orienté qui représente les différents régimes alimentaires et leurs incompatibilités. Nous ajoutons des arêtes pour chaque paire de régimes alimentaires compatibles, puis nous utilisons la méthode `greedy_color()` pour trouver la coloration optimale du graphe. Le nombre de couleurs utilisées pour colorer le graphe correspond au nombre minimum de plats nécessaires pour satisfaire tous les régimes alimentaires compte tenu des incompatibilités.

Nous avons également utilisé la bibliothèque ***Matplotlib*** de Python pour visualiser notre graphe et sa coloration. Matplotlib est une bibliothèque Python de traçage de graphiques en 2D, qui nous a permis de représenter notre graphe avec des couleurs différentes pour chaque régime alimentaire.

<h2 style="text-align:center; color:orange">III. La résolution du problème</h2>

**1. Importation des librairies : nous utiliserons la librairie `networkx` pour la manipulation de graphes et matplotlib pour la visualisation.**

```python
import networkx as nx
import matplotlib.pyplot as plt
```

**2. Création d'un graphe non orienté vide : nous créons un graphe non orienté vide pour représenter les relations entre les différents régimes alimentaires :**

```python
G = nx.Graph()
```

**3. Ajout des nœuds représentant les 8 régimes alimentaires au graphe :**

```python
regimes_alimentaires = ['Sans gluten', 'Végétalien', 'Sans lactose', 'Sans fruits à coque', 'Sans produits laitiers', 'Sans œufs', 'Sans poisson', 'Sans viande rouge']
G.add_nodes_from(regimes_alimentaires)
print("Nos différents régimes alimentaires : ",G.nodes())
```

**4. Affichage du graphe : nous visualisons le graphe pour voir les nœuds représentant les régimes alimentaires.**

```python
# Spécifier les couleurs pour chaque régime alimentaire
colors = ['blue', 'green', 'red', 'purple', 'orange', 'brown', 'pink', 'gray']

# Traçage du graphe avec les couleurs spécifiées pour chaque nœud
plt.figure(figsize=(5, 5)) 
nx.draw(G, with_labels=True, node_color=colors, node_size=500, font_size=9, font_weight='bold')
plt.show()
```

**5.Ajout des arêtes pour les paires de régimes alimentaires qui ne sont pas en conflit :**

```python
edges_alimentaires = [('Sans gluten', 'Sans lactose'), 
                      ('Sans gluten', 'Sans fruits à coque'), 
                      ('Sans gluten', 'Sans produits laitiers'), 
                      ('Végétalien', 'Sans œufs'), 
                      ('Végétalien', 'Sans poisson'), 
                      ('Végétalien', 'Sans viande rouge'), 
                      # ('Sans lactose', 'Sans produits laitiers'), 
                      ('Sans lactose', 'Sans œufs'), 
                      ('Sans fruits à coque', 'Sans produits laitiers'), 
                      ('Sans fruits à coque', 'Sans œufs'), 
                      ('Sans fruits à coque', 'Sans poisson'), 
                      ('Sans fruits à coque', 'Sans viande rouge'), 
                      ('Sans produits laitiers', 'Sans œufs'), 
                      ('Sans produits laitiers', 'Sans poisson'), 
                      ('Sans produits laitiers', 'Sans viande rouge'), 
                      ('Sans œufs', 'Sans poisson'), 
                      ('Sans œufs', 'Sans viande rouge'), 
                      ('Sans poisson', 'Sans viande rouge')]

# Ajouter les arêtes au graphe
G.add_edges_from(edges_alimentaires)

# Afficher le graphe
plt.figure(figsize=(7, 7)) 
nx.draw(G, with_labels=True, node_size=500, font_size=9, font_weight='bold')
plt.show()
```

En ajoutant ces arêtes entre les régimes alimentaires compatibles, nous avons créé un graphe qui représente les relations entre les différents régimes alimentaires.

**6. Trouver la coloration optimale du graphe :**

```python
# Trouver la coloration optimale du graphe
colors_alimentaires = nx.greedy_color(G)

# Afficher le graphe avec les couleurs optimales
color_map_alimentaire = [colors_alimentaires.get(node) for node in G.nodes()]
plt.figure(figsize=(7, 7))
nx.draw(G, with_labels=True, node_color=color_map_alimentaire, node_size=500, font_size=9, font_weight='bold')
plt.show()
```

En utilisant la fonction ``nx.greedy_color()``, nous trouvons la coloration optimale du graphe représentant les régimes alimentaires compatibles. Ensuite, nous utilisons ces couleurs pour afficher le graphe colorié, où les régimes alimentaires compatibles sont reliés par des arêtes. Cette représentation nous permet de visualiser facilement les groupes de régimes alimentaires qui peuvent coexister dans le même menu sans entraîner de conflits alimentaires.

```python
print(colors_alimentaires)
```

Les couleurs attribuées aux régimes alimentaires par l'algorithme de coloration peuvent varier en fonction de l'exécution de l'algorithme. Cependant, chaque couleur représente un groupe de régimes alimentaires qui peuvent coexister sans entraîner de conflits alimentaires. Si on souhaite connaître les couleurs spécifiques attribuées à chaque régime alimentaire, on peut afficher le dictionnaire `colors_alimentaires` retourné par la fonction `nx.greedy_color()`. 

Cela affichera un dictionnaire où les clés sont les régimes alimentaires et les valeurs sont les couleurs attribuées à chaque régime alimentaire. On voit ainsi quel régime alimentaire est associé à quelle couleur dans la coloration optimale du graphe.

```python
display(color_map_alimentaire)
```

Le `color_map_alimentaire` est une liste qui associe chaque nœud du graphe à sa couleur respective dans la visualisation colorée du graphe. Chaque élément de la liste représente la couleur attribuée à un nœud spécifique du graphe.

Cela affichera la liste des couleurs attribuées à chaque nœud du graphe, dans l'ordre des nœuds. On peut voir quelle couleur est associée à chaque régime alimentaire dans la visualisation colorée du graphe.

```python
print(G.nodes())
```

**7. Affichage du nombre minimum de plats nécessaires :**

```python
num_plats = max(colors_alimentaires.values()) + 1
print("Le nombre minimum de plats nécessaires pour les 8 différentes régimes compte tenu des incompatibilités est de ", num_plats)
```

Dans ce contexte, chaque couleur représente un groupe de régimes alimentaires compatibles, et le nombre de couleurs correspond au nombre minimum de plats nécessaires pour couvrir tous les régimes alimentaires sans conflits. En affichant le nombre de couleurs, nous obtenons donc le nombre minimum de plats nécessaires pour satisfaire tous les régimes alimentaires sans qu'ils ne se chevauchent.
